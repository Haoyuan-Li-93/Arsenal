```python
import numpy as np
import skimage.measure as skm
import matplotlib.pyplot as plt
```

# Load mask and downsample the mask

```python
mask_2d = np.load('/reg/d/psdm/amo/amo86615/scratch/haoyuan/psocake/r0186/mask_2D.npy')
mask_ds = skm.block_reduce(block_size=(4, 4), func=np.min, image=mask_2d)
```

```python
fig = plt.figure(figsize=(16, 12))
plt.imshow(mask_ds)
plt.colorbar()
plt.show()
```

# Save the downsampled mask

```python
np.save('../input/mask.npy',mask_ds)
```
